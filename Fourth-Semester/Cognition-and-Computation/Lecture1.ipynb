## **The Computational Nature of Cognition**
### **Core Idea (Paul Thagard, 2005)**
- **Thinking = Representational structures in mind + Computational procedures**
- **Mind as an Information Processing System**  
  - **Input (stimulus) → Processing (cognition) → Output (response)**

### **Historical Perspectives**
- **Chomsky (1959):** Criticized behaviorism → Led to cognitive revolution.
- **Anderson (1980):** A cognitive theory should function like a computer program (explicit assumptions/variables).

### **CRUM (Computational-Representational Understanding of Mind)**
- Understanding cognition in terms of **representations + computations**.
- Thinking = **Mental representations + Computational procedures**  
  *(Similar to how programs = Data structures + Algorithms)*

## **Reasoning and Problem Solving**
### **Planning (Logic-Based Approach)**
- Uses **search algorithms** to find solutions.
- Example: **Newell & Simon's General Problem Solver (1959)**
  - Algorithmic problem solver based on **search**.

### **Problem Solving:**
- A problem exists when:
  - **Current state ≠ Goal state**
  - **Path to goal is not obvious**
  
### **Means–End Analysis (Strategy for Problem Solving)**
- Reduce gap between **current state** & **goal state** by setting subgoals.
- **Operators** used:
  - **Move**
  - **Expand**
  - **Delete**
  
#### **Example: Tower of Hanoi**
- **Rules:**
  - Move all discs from **peg A → peg C**.
  - Only one disc can move at a time.
  - Larger discs cannot be placed on smaller discs.
  
### **State Space & Action Space**
- **State Space:** All possible states the system can be in.
- **Action Space:** All possible actions to move between states.

## **Perceptual Decision Making**
### **What is Perceptual Decision Making?**
- Choosing between **alternatives** based on **sensory evidence**.
- **Example:** Determining if a tennis ball was in or out.

### **Modeling Decision Making**
- **Neural Accumulators:** Each choice has a **corresponding accumulator** that gathers evidence over time.
- **Decision made when threshold is reached.**

#### **Race Models**
- Each choice alternative has an **accumulator**.
- All accumulators **race toward a decision threshold**.
- Once an accumulator reaches the threshold → **Decision is made**.

## **Language**
### **The Word Segmentation Problem**
- When exposed to a new language, speech seems like **a stream of random sounds**.
- **Infant-directed speech:** Helps infants learn language through exaggerated intonation & stress.
- **Statistical Learning:**
  - Transitional probabilities between phonemes help identify **word boundaries**.
  - **Low transitional probability = Likely word boundary**.

## **Connectionism (Neural Networks & Cognitive Science)**
### **Revival of Connectionism (1980s)**
- AI winter (1970s) → **Rumelhart & McClelland revived connectionism**.
- **Parallel Distributed Processing (PDP):** Inspired by neural networks.

### **Key Features of Connectionist Models**
- **Biologically Inspired:** Modeled after the human brain.
- **Lesion Tolerant:** Can function even if parts are damaged.
- **Generalization:** Can apply learned rules to **new inputs**.

### **Neural Structure**
- **Neurons receive input** through **dendrites**.
- **Neurons send output** through **axons**.
- **Massive connectivity:** 20 billion neurons, 15 × 10¹³ synapses.

### **McClelland’s 1981 Memory Model**
- **Memory is content-addressable**.
- Stored in **connections between neurons, not individual neurons**.
- **Excitatory & Inhibitory connections** contribute to information processing.
